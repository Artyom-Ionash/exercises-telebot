## Продолжаем работу в файле main.py

Выполняем парсинг страниц с цитатами персонажей.


```python
# Импортируем необходимые библиотеки и модули
import pickle
from bs4 import BeautifulSoup
from time import sleep
from selenium import webdriver
from selenium.webdriver.remote.webdriver import WebDriver

# Задаём переменную data, содержащую пары "категория-адрес"
data = [
    ["Крош", "https://citaty.info/character/krosh"],
    ["Ёжик", "https://citaty.info/character/ezhik"],
    ["Кар-Карыч", "https://citaty.info/character/kar-karych-0"],
    ["Нюша", "https://citaty.info/character/nyusha"],
]


def parse_page(soup: BeautifulSoup):
    """
    Находит все элементы <div> с классом "post noSidePadding" в объекте BeautifulSoup,
    а затем извлекает текст из элементов с тегом <div>.
    """
    quotes_list = []
    posts = soup.find_all("a", class_="citatyinfo-quote alink")
    for post in posts:
        quotes_list.append(post.text.strip())
    return quotes_list


def cook_soup(driver: WebDriver):
    """ """
    return BeautifulSoup(driver.page_source, features="lxml")


# Создаём объект драйвера Chrome
driver = webdriver.Chrome()
quotes: dict[str, list[str]] = {}
# Проходимся в цикле по списку data
for category, url in data:
    # Для каждой категории и URL-адреса открываем страницу и ожидаем 3 секунды
    driver.get(url)
    sleep(3)

    # Извлекаем анекдоты с текущей страницы, а результаты сохраняем в словарь quotes
    soup = cook_soup(driver)
    quotes[category] = parse_page(soup)

# Словарь сохраняем в файл quotes.pickle
with open("quotes.pickle", "wb") as f:
    pickle.dump(quotes, f)
```

### Визуализируем результат


```python
import json

print(json.dumps(quotes, indent=4, ensure_ascii=False))
```
